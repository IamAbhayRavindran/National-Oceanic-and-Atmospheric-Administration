# ***Avoided Leap Year February 29th ***





```python
file_name = next(iter(uploaded))
df = pd.read_csv(io.BytesIO(uploaded[file_name]))

df['Date'] = pd.to_datetime(df['Date'])

df['is_leap_year'] = df['Date'].dt.is_leap_year

df = df[~((df['Date'].dt.month == 2) & (df['Date'].dt.day == 29) & (df['is_leap_year']))]

df = df.drop(columns=['is_leap_year'])
```

# ***Daily High & Low Temperature ( 2005 - 2014 ) ***

```python
df['Date'] = pd.to_datetime(df['Date'])

print(df.columns)

high_temp_col = 'Data_Value'
low_temp_col = 'Data_Value'

df = df[(df['Date'].dt.year >= 2005) & (df['Date'].dt.year <= 2014)]

daily_df = df.groupby(df['Date'].dt.date).agg({high_temp_col: 'mean', low_temp_col: 'mean'})


plt.figure(figsize=(12, 6))

plt.plot(daily_df.index, daily_df[high_temp_col], label='High Temperature', color='green')
plt.plot(daily_df.index, daily_df[low_temp_col], label='Low Temperature', color='yellow')
plt.gca().set_facecolor('black')
plt.gcf().set_facecolor('red')

plt.fill_between(daily_df.index, daily_df[high_temp_col], daily_df[low_temp_col], color='gray', alpha=0.3)

plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Daily High and Low Temperatures (2005-2014)')
plt.legend()
plt.grid(True)
plt.show()
```

# **Overlay 2015 Datas when the temperature High and Low ( 2005 - 2014 )**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = df[~((df['Date'].dt.month == 2) & (df['Date'].dt.day == 29) & (df['Date'].dt.is_leap_year))]

df_2005_2014 = df[(df['Date'].dt.year >= 2005) & (df['Date'].dt.year <= 2014)]
df_2015 = df[df['Date'].dt.year == 2015]

record_highs = df_2005_2014.groupby(df_2005_2014['Date'].dt.strftime('%m-%d'))['Data_Value'].max()
record_lows = df_2005_2014.groupby(df_2005_2014['Date'].dt.strftime('%m-%d'))['Data_Value'].min()

record_breaking_highs = df_2015[df_2015.apply(lambda row: row['Data_Value'] > record_highs.get(row['Date'].strftime('%m-%d'), -np.inf), axis=1)]
record_breaking_lows = df_2015[df_2015.apply(lambda row: row['Data_Value'] < record_lows.get(row['Date'].strftime('%m-%d'), np.inf), axis=1)]

plt.figure(figsize=(12, 6))

plt.scatter(df_2005_2014['Date'].dt.dayofyear, df_2005_2014['Data_Value'], color='orange', alpha=0.1, label='2005-2014',s=10)

plt.scatter(df_2015['Date'].dt.dayofyear, df_2015['Data_Value'], color='blue', label='2015')

plt.scatter(record_breaking_highs['Date'].dt.dayofyear, record_breaking_highs['Data_Value'], color='red', label='Record Highs (2015)', edgecolor='black', s=10)

plt.scatter(record_breaking_lows['Date'].dt.dayofyear, record_breaking_lows['Data_Value'], color='purple', label='Record Lows (2015)', edgecolor='black', s=10)
plt.gca().set_facecolor('black')
plt.gcf().set_facecolor('red')

plt.xlabel('Day of Year')
plt.ylabel('Temperature')
plt.title('Record-Breaking Temperatures in 2015 Compared to 2005-2014')
plt.legend()
plt.grid(True)

plt.show()
```

# Legends, Labels & chart Junk

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)


fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(x, y1, label='Sine Wave', linestyle='-', linewidth=2, color='blue')
ax.plot(x, y2, label='Cosine Wave', linestyle='--', linewidth=2, color='red')

ax.set_xlabel('X-axis (Time in seconds)', fontsize=12)
ax.set_ylabel('Y-axis (Amplitude)', fontsize=12)
ax.set_title('Sine and Cosine Waves', fontsize=14, fontweight='bold')


ax.legend(loc='upper right', fontsize=10, frameon=False)


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(True, linestyle='--', alpha=0.6)

plt.gca().set_facecolor('black')
plt.gcf().set_facecolor('yellow')

plt.show()
```

# ** Map : Ann Arbor, Michigan, United States **

```python
import folium


ann_arbor_coords = [42.2808, -83.7430]
michigan_coords = [44.3148, -85.6024]
us_coords = [37.0902, -95.7129]


m = folium.Map(location=us_coords, zoom_start=4, tiles="Esri.WorldImagery")


locations = {
    "Ann Arbor, MI": (ann_arbor_coords, "blue"),
    "Michigan": (michigan_coords, "green"),
    "United States": (us_coords, "red"),
}

for name, (coords, color) in locations.items():
    folium.Marker(
        coords,
        popup=name,
        tooltip=f"Click for {name}",
        icon=folium.Icon(color=color),
    ).add_to(m)


m
```

# **# Temperature Ann Arban, Michigan, US**

```python

months = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]


data = {
    'Month': months,
    'Ann Arbor, MI': [23.1, 21.6, 33.9, 47.5, 62.1, 69.9, 72.8, 70.8, 65.0, 52.1, 42.3, 31.2],
    'United States': [38.0, 35.9, 45.4, 53.2, 60.8, 71.8, 74.9, 73.9, 68.5, 57.4, 48.0, 38.6]
}


df = pd.DataFrame(data)
df.head()
```

```python
plt.figure(figsize=(12, 6))


plt.plot(df['Month'], df['Ann Arbor, MI'], label='Ann Arbor, MI', color='blue', marker='o')
plt.plot(df['Month'], df['United States'], label='United States', color='red', marker='o')


plt.title('Monthly Average Temperatures in 2015', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Temperature (°F)', fontsize=12)


plt.legend()

plt.xticks(rotation=45)


plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().set_facecolor('black')
plt.gcf().set_facecolor('green')


plt.tight_layout()
plt.show()
```
